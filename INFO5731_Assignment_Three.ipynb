# **Question 1: Understand N-gram**

(45 points). Write a python program to conduct N-gram analysis based on the dataset in your assignment two:

(1) Count the frequency of all the N-grams (N=3).

(2) Calculate the probabilities for all the bigrams in the dataset by using the fomular count(w2 w1) / count(w2). For example, count(really like) / count(really) = 1 / 3 = 0.33.

(3) Extract all the **noun phrases** and calculate the relative probabilities of each review in terms of other reviews (abstracts, or tweets) by using the fomular frequency (noun phrase) / max frequency (noun phrase) on the whole dataset. Print out the result in a table with column name the all the noun phrases and row name as all the 100 reviews (abstracts, or tweets). 


```
import requests
from bs4 import BeautifulSoup
import pandas as pd
# Title
main_text = [] 
# Reviews
sub_text =[] 
for number in range(52):
  link = "https://www.imdb.com/title/tt7286456/reviews?ref_=tt_ql_3" + str(number)
  page = requests.get(link) 
  soup = BeautifulSoup(page.text, 'html.parser')
  main_reviews = soup.find_all(class_='title') 
  text_reviews = soup.find_all(class_='text show-more__control')
  for ele, sub_ele in zip(main_reviews, text_reviews) :
      main_text.append(ele.text) 
      sub_text.append(sub_ele.text)
 # Creating Dataframe
df = pd.DataFrame(list(zip(main_text, sub_text)), columns =['Title', 'Review']) 
df['Word_count'] = df['Review'].apply(lambda x:len(str(x).split(" ")))
df.head(100)
```

```
#Data cleaning
#Punctuation removal
df['Review'] = df['Review'].str.replace('[^\w\s]','')  
#Lower case
df['Review'] = df['Review'].apply(lambda x: " ".join(x.lower() for x in x.split()))
#Stopwords removal
from nltk.corpus import stopwords
nltk.download('stopwords')
stop = stopwords.words('english')
df['Review'] = df['Review'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
#converting to a file
data = ''
for i in range(0,len(df['Review'])):
  data+= df['Review'][i]
print(data)
```

```
# Trigrams
nltk.download('punkt')
words = nltk.word_tokenize(data)
result_trigrams = nltk.trigrams(words)
freq_trigrams = nltk.FreqDist(result_trigrams)
freq_trigrams
```

```
# Bigrams
result_bigrams = nltk.bigrams(words)
import re
from itertools import tee, islice
import collections
def ngrams(lst, n):
  tlst = lst
  while True:
    a, b = tee(tlst)
    l = tuple(islice(a, n))
    if len(l) == n:
      yield l
      next(b)
      tlst = b
    else:
      break
data2 = collections.Counter(ngrams(data.split(), 2))
print(data2)
```

```
# Joining bigrams
s1 = ''
for w1,w2 in result_bigrams:
  s1 = s1 + ' ' + w1 + ' ' + w2
list = s1.split()
print(list)
```

```
# Probability
for w1,w2 in data2:  
  den = list.count(w1)
  num = data2[(w1,w2)]
  if den == 0:
    break
  prob = num/den
  print(w1,w2,prob)
```

```
 # Noun phrase
from collections import Counter 
import nltk
from nltk import word_tokenize, pos_tag, pos_tag_sents
nltk.download('averaged_perceptron_tagger')
lower_case = data.lower()
tokens = nltk.word_tokenize(lower_case)
tags = nltk.pos_tag(tokens)
counts = Counter( tag for word,  tag in tags)
highest_noun_count = counts['NN']
df['phrases'] = pos_tag_sents(df['Review'].apply(word_tokenize).tolist())
y = df['phrases']
list1 = []
list2 = []
for i in y:
  count = Counter( tag for word,  tag in i)
  list1.append(count['NN'])
for i in list1:
  j = i/highest_noun_count
  list2.append(j)
print(list2)
df1 = pd.DataFrame(list1,columns = ['noun_count'])
df2 = pd.DataFrame(list2,columns = ['prob_noun'])
result = pd.concat([df1,df2], axis=1, sort=False)
finalresult = pd.concat([df,result], axis = 1, sort = False)
finalresult.head(100)
```

```
import nltk  #cleaning a single sentence
nltk.download('stopwords')
nltk.download('wordnet')
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
query = "While it succeeds in aping the superficial aspects of [Martin] Scorsese's style, Joker lacks the intelligence or gravitas of Scorsese's best work"
lemmatizer = WordNetLemmatizer()
regex = re.compile('[^a-zA-Z\s]')

def cleaningdata(text):
  text =  re.sub(regex, '', text) # Removes Special characters 
  words = " ".join([word for word in text.split() if word not in stopwords.words('english')]) # removes stopwords from text
  lemmatize =   " ".join([lemmatizer.lemmatize(i) for i in words.lower().split()]) # makes word in to meaning
  return lemmatize
cleaned_query = cleaningdata(query)
print('Cleaned Query: ' + cleaned_query)
```

```
#cosine similarity
import pandas
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
text = []
for x in df['Review']:
  text.append(x)

text.append(cleaned_query)
vectorizer = TfidfVectorizer(input=text, min_df=0.1, stop_words='english')
matrix = vectorizer.fit_transform(text)

cs_dict = {}
for i in range(0, matrix.shape[0]):
  sim=cosine_similarity(matrix[80], matrix[i])
  cs_dict[str(i)] = sim[0][0]
  i+=1

sorted_keys = (sorted(cs_dict, key=cs_dict.get,reverse=True))
sorted_cosine_dict = {}
for key in sorted_keys:
  sorted_cosine_dict[key] = cs_dict[key];

review = []
cos = []
corpus =[]
for i in text:
  corpus.append(i)
for y in sorted_cosine_dict:
  review.append(corpus[int(y)])
  cos.append(sorted_cosine_dict[y])
df1 = pd.DataFrame(review, columns = ['Review'])

df1['Rank'] = [i for i in range(1, len(sorted_keys)+1)]
df1['Cosine_similarity'] = cos
df1.head(1)
```
